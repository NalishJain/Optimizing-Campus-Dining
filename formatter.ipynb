```python
months = [
#     'Sep',
    # 'Oct', 'Nov', 'Dec',
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug']

def year(month):
    return 2023 if month in ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug'] else 2022
```

```python
df = pd.DataFrame()
for month in months:
    df = pd.concat([df, pd.read_excel(f'./data/excel_files/Formatted{month}{year(month)}.xlsx')], ignore_index=True)

df
```

```python
from datetime import datetime

summer_start = datetime.strptime("2023-05-15", "%Y-%m-%d")
summer_end = datetime.strptime("2023-08-07", "%Y-%m-%d")

# drop all rows where the date is NaT
df = df.dropna(subset=['Date'])

dates = df['Date'].tolist()
weekdays = list()
for i in range(len(dates)):
    try:
        dates[i] = datetime.strptime(str(dates[i]).split(' ')[0], "%Y-%m-%d")
    except:
        dates[i] = datetime.strptime(str(dates[i]).split(' ')[0], "%m/%d/%Y")

    weekdays.append(dates[i].weekday())

semtypes = list()
for i in range(len(dates)):
    semtypes.append(dates[i] > summer_start and dates[i] < summer_end)

df['SemType'] = semtypes

# Weekday format: 6 for Sunday, 0 for Monday
df['Weekday'] = weekdays

holidays_list = []
holidays = list()
for i in range(len(dates)):
    holidays.append(dates[i] in holidays_list)

df.loc[:, 'Holiday'] = holidays

df = df.iloc[:, [0, 7, 1, 2, 3, 4, 5, 6]]
df.head()
```

```python

all_months_menu_dict = {
    'Jan': jan_menu_dict,
    'Feb': feb_menu_dict,
    'Mar': mar_menu_dict,
    'Apr': apr_menu_dict,
    'May': may_menu_dict,
    'Jun': jun_menu_dict,
    'Jul': jul_menu_dict,
    'Aug': aug_menu_dict,
}

import json
json_menu = {}
for month, menu_dict in all_months_menu_dict.items():
    json_menu[month] = [{} for _ in range(7)]
    for (day, meal), items in menu_dict.items():
        json_menu[month][int(day)][meal] = items
        

# write the all_months_menu_dict to a json file
# with open('data/month_wise_menu.json', 'w') as fp:
#     json.dump(json_menu, fp)

# Encode the menus into vectors
global_menu_map = {}

for month, menu_dict in all_months_menu_dict.items():
    for day_and_meal, items in menu_dict.items():
        for item in items:
            if item not in global_menu_map:
                global_menu_map[item] = len(global_menu_map)

# Encode the menus into vectors
global_vector_map = {}

for month, menu_dict in all_months_menu_dict.items():
    for day, meal in menu_dict.keys():
        menu = menu_dict[(day, meal)]
        menu_vector = np.zeros(len(global_menu_map.keys()))
        for item in menu:
            menu_vector[global_menu_map[item]] = 1
        global_vector_map[(month, day, meal)] = menu_vector


print("Number of unique meals we have:", len(global_vector_map.keys()))
print("Number of unique items:", len(list(global_vector_map.values())[0]))
print(global_menu_map)
```
