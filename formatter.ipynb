```python
global_menu_map = {}
counter = 0
for day_and_meal in jan_menu_dict.keys():
    for item in jan_menu_dict[day_and_meal]:
        if item not in global_menu_map.keys():
            global_menu_map[item] = counter
            counter += 1

from pprint import pprint as prettyprint
prettyprint(global_menu_map.values())
```

```python

all_months_menu_dict = {
    'Jan': jan_menu_dict,
    'Feb': feb_menu_dict,
    'March': mar_menu_dict,
    'April': apr_menu_dict,
    'May': may_menu_dict,
    'June': jun_menu_dict,
    'July': jul_menu_dict,
    'August': aug_menu_dict,
}

import json
json_menu = {}
for month, menu_dict in all_months_menu_dict.items():
    json_menu[month] = [{} for _ in range(7)]
    for (day, meal), items in menu_dict.items():
        json_menu[month][int(day)][meal] = items
        

# write the all_months_menu_dict to a json file
with open('data/month_wise_menu.json', 'w') as fp:
    json.dump(json_menu, fp)

# Encode the menus into vectors
global_menu_map = {}

for month, menu_dict in all_months_menu_dict.items():
    for day_and_meal, items in menu_dict.items():
        for item in items:
            if item not in global_menu_map:
                global_menu_map[item] = len(global_menu_map)

# Encode the menus into vectors
global_vector_map = {}

for month, menu_dict in all_months_menu_dict.items():
    for day, meal in menu_dict.keys():
        menu = menu_dict[(day, meal)]
        menu_vector = np.zeros(len(global_menu_map.keys()))
        for item in menu:
            menu_vector[global_menu_map[item]] = 1
        global_vector_map[(month, day, meal)] = menu_vector


print("Number of unique meals we have:", len(global_vector_map.keys()))
print("Number of unique items:", len(list(global_vector_map.values())[0]))
```
