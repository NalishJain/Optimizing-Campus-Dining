```python

all_months_menu_dict = {
    2022: {
        'Sep': sep_menu_dict,
        'Oct': oct_menu_dict,
        'Nov': nov_menu_dict,
        'Dec': dec_menu_dict,
    },

    2023: {
        'Jan': jan_menu_dict,
        'Feb': feb_menu_dict,
        'Mar': mar_menu_dict,
        'Apr': apr_menu_dict,
        'May': may_menu_dict,
        'Jun': jun_menu_dict,
        'Jul': jul_menu_dict,
        'Aug': aug_menu_dict,
        'Sep': sep_23_menu_dict,
    }
}

import json
json_menu = {}
for year in all_months_menu_dict.keys():
    json_menu[year] = {}
    for month, menu_dict in all_months_menu_dict[year].items():
        json_menu[year][month] = [{} for _ in range(7)]
        for (day, meal), items in menu_dict.items():
            json_menu[year][month][int(day)][meal] = items
        

# write the all_months_menu_dict to a json file
# with open('data/month_wise_menu.json', 'w') as fp:
#     json.dump(json_menu, fp)

# Encode the menus into vectors
global_menu_map = {}

for year in all_months_menu_dict.keys():
    for month, menu_dict in all_months_menu_dict[year].items():
        for day_and_meal, items in menu_dict.items():
            for item in items:
                if item not in global_menu_map:
                    global_menu_map[item] = len(global_menu_map)

# Encode the menus into vectors
global_vector_map = {}

for year in all_months_menu_dict.keys():
    for month, menu_dict in all_months_menu_dict[year].items():
        for day, meal in menu_dict.keys():
            menu = menu_dict[(day, meal)]
            menu_vector = np.zeros(len(global_menu_map.keys()))
            for item in menu:
                menu_vector[global_menu_map[item]] = 1
            global_vector_map[(year, month, day, meal)] = menu_vector


print("Number of unique meals we have:", len(global_vector_map.keys()))
print("Number of unique items:", len(list(global_vector_map.values())[0]))
print(global_menu_map)

print(json_menu)
```
